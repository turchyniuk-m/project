###Sales by device type and model (%)

```python
device_sales = df.groupby(['device','mobile_model_name'])['price'].sum()
device_sales_pct = (device_sales / df['price'].sum() * 100).sort_values(ascending=False)
print("\nSales by device type and model (%) / Продажі за типами та моделями девайсів (%):")
print(device_sales_pct.head(20))
```

```python
df_bar = device_sales_pct.reset_index()
df_bar.columns = ['device', 'model', 'sales_pct']


fig = px.bar(
    df_bar.head(20),
    y='device',
    x='sales_pct',
    color='model',
    orientation='h',

    text='model',
    title='Sales by device type and model (%)'
)


fig.update_traces(
    texttemplate="<b>%{text}</b><br>%{x:.1f}%",
    textposition='inside',
    insidetextanchor='middle',
    textfont=dict(size=12, color='white')
)


fig.update_layout(
    paper_bgcolor='black',
    plot_bgcolor='black',
    font_color='white',
    showlegend=False,
    xaxis=dict(
        showgrid=False,
        zeroline=False,
        visible=False
    ),
    yaxis=dict(
        showgrid=False,
        title=""
    ),
    height=500
)

fig.update_traces(marker_line_width=0)

fig.show()
```

###Sales by traffic source (%)

```python
traffic_sales_pct = (df.groupby('traffic_source')['price'].sum() / df['price'].sum() * 100).sort_values(ascending=False)
print("\nSales by traffic source (%):")
print(traffic_sales_pct)
```

```python
labels = traffic_sales_pct.index
values = traffic_sales_pct.values

colors = plt.cm.viridis(np.linspace(0, 1, len(labels)))

plt.figure(figsize=(10, 7), facecolor='black')
ax = plt.gca()
ax.set_facecolor('black')

wedges, texts, autotexts = ax.pie(
    values,
    labels=labels,
    autopct='%1.1f%%',
    startangle=140,
    colors=colors,
    pctdistance=0.85,
    textprops={'color': 'white', 'weight': 'bold'}
)

for autotext in autotexts:
    autotext.set_color('black')
    autotext.set_fontsize(10)

plt.legend(wedges, labels,
          title="Sources",
          loc="center left",
          bbox_to_anchor=(1, 0, 0.5, 1),
          facecolor='#222222',
          edgecolor='white',
          labelcolor='white')

plt.title('Sales by Traffic Source (%)', color='white', fontsize=18, fontweight='bold', pad=20)

plt.axis('equal')
plt.tight_layout()
plt.show()
```

###Countries with most registered users



```python
users_by_country = df.groupby('country')['account_id'].nunique().sort_values(ascending=False)
print("\nCountries with most registered users:")
print(users_by_country.head(10))
```

```python
top_n = 5
data_to_plot = users_by_country.head(top_n)
other_sum = users_by_country.iloc[top_n:].sum()

final_data = pd.concat([data_to_plot, pd.Series({'Other': other_sum})])

plt.figure(figsize=(10, 7), facecolor='black')
ax = plt.gca()
ax.set_facecolor('black')


colors = plt.cm.Paired(np.linspace(0, 1, len(final_data)))

wedges, texts, autotexts = ax.pie(
    final_data.values,
    labels=final_data.index,
    autopct='%1.1f%%',
    startangle=140,
    colors=colors,
    pctdistance=0.8,
    explode=[0.05] * len(final_data),
    textprops={'color': 'white', 'weight': 'bold'}
)

plt.setp(autotexts, size=10, weight="bold", color="black")

plt.legend(
    wedges, final_data.index,
    title="Countries",
    loc="center left",
    bbox_to_anchor=(1, 0, 0.5, 1),
    facecolor='#222222',
    edgecolor='white',
    labelcolor='white'
)

plt.title('Countries with most registered users (%)', color='white', fontsize=18, fontweight='bold', pad=20)

plt.axis('equal')
plt.tight_layout()
plt.show()
```
